```python
# Get Ugandan maize farm observations
uganda_maize = df[
    (df['country'] == 'Uganda') &
    (df['crop'] == 'maize') 
]
total_farm_observations = len(uganda_maize)
print(f"Farm observations from Uganda on maize:", total_farm_observations)
```

```python
# Find Ugandan districts from which the data was collected from
print(list(uganda_maize['district'].unique())) 
```

```python
# Results show Kamuli, Iganga and Jinja which are all found in Eastern Uganda
```

```python
# How many years were considered during the study
print(list(uganda_maize['year'].unique())) 
```

```python
uganda_maize = df.loc[
    (df['country'] == 'Uganda') &
    (df['crop'] == 'maize') &
    (df['year'] >= 2015),
    ['unique_id', 'country', 'district', 'site', 'year', 'season', 'plant_date', 'harvest_date',
     'field_latitude', 'field_longitude', 'plot_acres', 'hybrid',
     'flood', 'drought', 'pest_disease', 'mlnd', 'msv', 'intercrop', 'dap_kg_pa', 'urea_kg_pa', 'weed', 'yield_kg_pa', 'yield_kg_ph']
].reset_index(drop=True)
print(uganda_maize.tail(10))
```

```python
# Calculate district-season median
district_season_median = uganda_maize.groupby(['district', 'season', 'year'])['yield_kg_pa'].transform('median')
uganda_maize['district_season_median_yield'] = district_season_median

# Calculate and apply stress threshold
uganda_maize['stress_threshold'] = 0.7 * district_season_median
uganda_maize['stress_label'] = (uganda_maize['yield_kg_pa'] < district_season_median).astype(int)
uganda_maize
```

```python
# Data quality checks

# Zero yields
print(f"Zero yield rows: ", (uganda_maize['yield_kg_pa'] == 0).sum())

# Any unreasonable high yields
print(f"Unreasonable high yields (>3000 kg/acre): ", (uganda_maize['yield_kg_pa'] > 3000).sum())

# Check GPS coordinate validity (Eastern Uganda bounds)
lat_ok = uganda_maize['field_latitude'].between(-2.0, 4.5)
lon_ok = uganda_maize['field_longitude'].between(32.0, 35.5)
print(f"Invalid farms (GPS): ", (~(lat_ok & lon_ok)).sum())

# Check for duplicate farm IDs
print("Duplicate farm IDs", uganda_maize['unique_id'].duplicated().sum())

# Check for missing values per column
print(uganda_maize.isnull().sum())
```

```python
# Count stressed and non-stressed farms
farms_with_stress = len(uganda_maize[uganda_maize['stress_label']==1])
farms_without_stress = len(uganda_maize[uganda_maize['stress_label']==0])

# Calculate percentages
stressed_farms_percent = (farms_with_stress / total_farm_observations) * 100
nonstressed_farms_percent = (farms_without_stress / total_farm_observations) * 100

print(f"Stressed farms: {farms_with_stress} => {stressed_farms_percent:.2f}%")
print(f"Non-stressed farms: {farms_without_stress} => {nonstressed_farms_percent:.2f}%")
```

```python
# Function to determine stress severity
def stress_severity(row):
    ratio = row['yield_kg_pa'] / row['district_season_median_yield']

    if ratio >= 0.7:
        return "None"
    elif ratio >= 0.5:
        return "Mild"
    elif ratio >= 0.3:
        return "Severe"
    else:
        return "Extreme"      
```

```python
uganda_maize['stress_severity'] = uganda_maize.apply(stress_severity, axis=1)
uganda_maize
```

```python
# Visualisation of the 256 farm observations
plt.figure(figsize=(8,5))
scatter = plt.scatter(uganda_maize['field_longitude'], uganda_maize['field_latitude'],
                     c=uganda_maize['stress_label'], cmap='RdYlGn_r', alpha=0.7, s=30)
plt.colorbar(scatter, label="Stress Label (1=Stressed)")
plt.title("Uganda Maize Farm Locations - One Acre Fund 2020")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
